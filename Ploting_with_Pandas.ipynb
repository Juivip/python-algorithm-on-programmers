```
import pandas as pd
import numpy as np
data = pd.DataFrame(np.random.rand(10, 4), columns=['A','B','C','D'])
data.head()
```

```
data.plot.bar()
data.plot.bar(stacked=True)
data.plot.barh(stacked=True)
data.plot.area()
```

```
data.plot.area(stacked=False,alpha=0.4)
```

```
data = pd.DataFrame(np.random.rand(100, 1), columns=['value']).reset_index()
data['value'].plot()
data['value'].rolling(10).mean().plot()
```

```
import pandas as pd
import numpy as np
data = pd.DataFrame(np.random.rand(10, 4), columns=['A','B','C','D'])
data.plot.kde(); #distribution plot
data.plot.scatter(x='A',y='B',c='C', s=data['C']*200) 
data.plot.hexbin(x='C',y='D', gridsize=18)
```
